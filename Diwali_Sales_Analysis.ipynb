```python
grouped_age_order=df.groupby(['Gender','Orders'])['User_ID'].agg('count').reset_index()
grouped_age_order_F=grouped_age_order.loc[grouped_age_order['Gender']=='F']
grouped_age_order_M=grouped_age_order.loc[grouped_age_order['Gender']=='M']
# grouped_age_order_F
```

# Exploratory Data Analysis

## Gender

```python
gender_count=df.groupby('Gender')['User_ID'].count().to_numpy()    #stores the total count of gender
sex=df['Gender'].drop_duplicates().to_numpy()    #stores the unique gender
```

```python
# plotting a bar chart for Gender and it's count

xpos=np.arange(len(sex))
plt.xticks(xpos,sex)
plt.ylabel('Count')
plt.xlabel('Sex')
plt.title('Bar Graph - Total_Count per Gender')
# plt.grid()
plt.bar(xpos,gender_count,label=['Female','Male'],color=['Red','Green'],width=0.5)
for i, count in enumerate(gender_count):
    plt.text(xpos[i], count, str(count), ha='center')
plt.legend()
plt.show()
```

```python
F_order=grouped_age_order_F['Orders'].to_list()
F_users=grouped_age_order_F['User_ID'].to_list()

M_order=grouped_age_order_M['Orders'].to_list()
M_users=grouped_age_order_M['User_ID'].to_list()

xposF=np.arange(len(F_order))
xposM=np.arange(len(M_order))
```

```python
#plotting graph for number of orders by different gender

plt.xticks(xposF,F_order)
plt.xticks(xposM,M_order)
plt.bar(xposF-0.2,F_users,width=0.4,label='Female')
plt.bar(xposM+0.2,M_users,width=0.4,label='Male')

for i,count in enumerate(M_users):
    plt.text(xposM[i]+0.2,count,str(count),ha='center',va='bottom')
for i,count in enumerate(F_users):
    plt.text(xposF[i]-0.2,count,str(count),ha='center',va='bottom')    

plt.xlabel('Order_Count')
plt.ylabel('Number of Orders')
plt.title('Bar Graph - Orders by different Gender')
plt.legend()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.title('Barplot - Total Amount per Gender')
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

## Age

```python
ax=sns.countplot(x=df['Age Group'],data=df,hue=df['Gender'])
for i in ax.containers:
    ax.bar_label(i)
    
plt.title('Countplot - Age group graph by gender')
```

```python
df.info()
```

```python
age_Torders=df.groupby('Age')['Orders'].sum().reset_index()
plt.plot(age_Torders.iloc[:,0],age_Torders.iloc[:,1],'g-.')
plt.xlabel('Age')
plt.ylabel('Total Orders')
plt.title('Line Graph - Total Order by Age')
plt.grid()
```

```python
age_Torders=df.groupby('Age Group')['Orders'].count().reset_index()
plt.scatter(age_Torders.iloc[:,0],age_Torders.iloc[:,1],color='#000000')
plt.xlabel('Age Group')
plt.ylabel('Order Count')
plt.title('Scatterplot - Total Order by Age Group')
plt.grid()
```

```python
age_Tamount=df.groupby('Age',as_index=False)['Amount'].sum().sort_values(by='Age')
plt.bar(age_Tamount.iloc[:,0],age_Tamount.iloc[:,1])
plt.xlabel('Age')
plt.ylabel('Total Amount spent')
plt.title('Bar Graph - Total Amount spent by Age')
plt.grid()
```

```python
age_Grp_Tamount=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y=age_Grp_Tamount.iloc[:,1],data=age_Grp_Tamount)
plt.xlabel('Age Groups')
plt.ylabel('Total Amount spent')
plt.title('Barplot Graph - Total Amount spent by Age Groups')
plt.grid()
```

*From the above Graphs it can be seen that people between age group 26-35 spend the most and the age group 45+ spend the least on shopping*

## State

```python
df.head()
```

```python
T_order_state=df.groupby(['State','Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending=False)
fig, ax = plt.subplots(figsize=(16, 10))
plt.xlabel('State')
plt.ylabel('order count')
plt.title('Total Orders by different state')
ax=sns.barplot(x='State',y='Orders',data=T_order_state,hue='Gender')
for i in ax.containers:
    ax.bar_label(i)

plt.xticks(rotation=90)
plt.show()
```

```python
T_Amount_state=df.groupby(['State','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
fig, ax = plt.subplots(figsize=(16, 10))
plt.xlabel('State')
plt.ylabel('Total Amount Spent')
plt.title('Total Amount spent by different state')
ax=sns.barplot(x='State',y='Amount',data=T_Amount_state,hue='Gender')
plt.xticks(rotation=90)
plt.show()
```

*From the above graphs it can be seen that the most sales are made in "Uttar Pradesh" followed by "Maharastra", However Punjab and Telengana purchased the least items*

## Marital Status

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
    
plt.title('Married/Unmarried Count')
plt.legend()
```

```python
M_Amount=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x=M_Amount['Marital_Status'],y=M_Amount['Amount'],data=M_Amount)
for i in ax.containers:
    ax.bar_label(i)
    
plt.title('Amount spent based on Marital Status')
```

*Unmarried people spend the most in shopping when compared to Married people*

## Occupations

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.xticks(rotation=90)
# ax=sns.countplot(x='Occupation',data=df)
# for i in ax.containers:
#     ax.bar_label(i)
    
count=df.groupby('Occupation')['User_ID'].count()
xpos=np.arange(len(df['Occupation'].unique()))

plt.xticks(xpos,df['Occupation'].unique())
plt.bar(xpos,count,color='#503525')
plt.xlabel('Occupations')
plt.ylabel('Count of Workers')
plt.title('Count of people with different profession')
for i,j in enumerate(count):
    plt.text(xpos[i],j,str(j),ha='center')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.xticks(rotation=90)
ax=df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x=ax['Occupation'],data=ax,y='Amount',hue='Gender')
plt.title('Total Spent on purchasing by different profession')
plt.show()
```

*From above graphs we can see that most of the buyers were from Aviation however people are working in IT & Healthcare spend more on purchasing when compared*

## Product Category

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax=sns.countplot(x='Product_Category',data=df)
plt.xticks(rotation=90)
for i in ax.containers:
    ax.bar_label(i)
plt.title('Total count of different product catogory')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.xticks(rotation=90)
ax=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=ax)
plt.title('Total Amount spent on different product category')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
